```python
#matplotlib的第一个学习案例画出简单图形
import matplotlib.pylab as plt
import numpy as np

x=np.linspace(-1,1,50)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
#matplotlib第二个案例，在每个figure中画图
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2
#在每个figure中显示图形，num给每个figure编号
plt.figure(num=1)
plt.plot(x,y1)

plt.figure(num=2)
plt.plot(x,y2)

plt.figure(num=3,figsize=(8,5))#设置figure的长为8宽为5
plt.plot(x,y1)
plt.plot(x,y2,color='red',linewidth=1.0,linestyle='--')#设置颜色，线的宽度，线的表现形式

plt.show()
```

```python
#matplotlib第三个案例，显示数学里面的文字单词以及希腊字母，图形的各种显示风格
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(30,100,5)
y=x
plt.figure(num=1,figsize=(8,5))
plt.plot(x,y,color='red',linestyle='--')

#取定xy轴的取值范围
plt.xlim(60,80)
plt.ylim(60,80)
#设置xy轴的标签
plt.xlabel('I am X')
plt.ylabel('I am Y')
#设置新下标
new_ticks=np.arange(60,100,10)
plt.xticks(new_ticks)
#plt.yticks([60,80,100],['normal','good','very good'])
#将字母转换为数学中的显示形式
plt.yticks([60,80,100],[r'$normal\alpha$',r'$good\beta$',r'$very\ good\gamma$'])

#利用正则表达式，其中希腊字母也可以表示出来
plt.show()
```

```python
#第四个案例调整坐标轴
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2
#在每个figure中显示图形，num给每个figure编号

plt.figure(num=3,figsize=(8,5))#设置figure的长为8宽为5
plt.plot(x,y1)
plt.plot(x,y2,color='red',linewidth=1.0,linestyle='--')#设置颜色，线的宽度，线的表现形式

plt.xlim(-1,2,5)
plt.ylim(-1,2,5)

plt.xlabel('I am X')
plt.ylabel('I am Y')

new_ticks=np.linspace(-1,2,5)
plt.xticks(new_ticks)
plt.yticks([-2,-1.8,-1,1.22,3],[r'$really\ bad$',r'$bad$',r'normal',r'$good$',r'$really\ good$'])

#gca get current axis获取当前框架

ax=plt.gca()

ax.spines['right'].set_color('none')#将右边框透明化
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')#用下面的坐标轴替换xaxis
ax.yaxis.set_ticks_position('left')
#挪动

ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
plt.show()
```
